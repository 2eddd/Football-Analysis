```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the data
table=pd.read_csv(r"C:\Users\Lenovo\Documents\Zahid\Python Projects\PL21-22 Table.csv")
table.head()
```

# Plotting a Bar Graph based on Points scored by each Club

```python
teamColours = ['#6CABDD','#C8102E','#034694','#DCDCDC',
              '#EF0107','#DA291C','#A52A2A','#003090',
               '#0057B8','#FDB913','#000000','#1B458F',
               '#C12026','#670E36','#D71920','#003399',
               '#1D428A','#6C1D45','#FBEE23','#00A650']
plt.bar(x=np.arange(1,21),height=table['Points'],color=teamColours)
plt.title("Premier League 2021-22")
plt.xticks(np.arange(1,21), table['Club'], rotation=90)

plt.xlabel("Team")
plt.ylabel("Points")
plt.show()
```

```python
#Create plot area
fig, ax = plt.subplots()

#Set plot size
fig.set_size_inches(7, 5)

#Plot chart as above, but change the plot type from 'o' to '*' - givng us stars!
plt.plot(table['GF'],table['GA'],"*")

#Add labels to chart area
ax.set_title("Goals for & Against")
ax.set_xlabel("Goals For")
ax.set_ylabel("Goals Against")

#Display the chart
plt.show()
```

```python
#Adding a crosshair to the chart to display the average line. This helps to see if a point is performing well or not.

fig, ax = plt.subplots()
fig.set_size_inches(7, 5)

plt.plot(table['GF'],table['GA'],"*")
plt.plot([table['GF'].mean(),table['GF'].mean()],[90,20],'k-', linestyle = ":", lw=1)
plt.plot([20,90],[table['GA'].mean(),table['GA'].mean()],'k-', linestyle = ":", lw=1)

ax.set_title("Goals for & Against")
ax.set_xlabel("Goals For")
ax.set_ylabel("Goals Against")

plt.show()
```

# Importing Data for the creation of Pie Chart

```python
#Creating a Pie Chart based on the CLub and their number of wins in the season
df=pd.DataFrame(table,columns=['Club','Won'])
df
```

```python
#Creating the Pie Chart
plt.pie(df['Won'])
plt.tight_layout()
```

# Labelling and Customizing the Pie Chart

```python
wp={'linewidth':1,'edgecolor':"black"}
values=df['Won']
fig,ax =plt.subplots(figsize=(12,8))
wedges=plt.pie(df['Won'],
       labels = df['Club'],
       colors=teamColours,
       startangle=90,
       shadow=True,)
plt.title(" No of Wins in Premier League 21/22")

plt.tight_layout()
```

# Adding Legend

```python
#Adding Legend to make it interpretable
values=df['Won']
colors=teamColours
labels=df['Club']
fig=plt.figure(figsize=(12,8))
plt.pie(values, colors=teamColours, labels=labels,
    autopct='%1.1f%%',shadow=True)
plt.title("No of Wins By Clubs in 21/22")
plt.legend(labels,title="Wins",loc="right",bbox_to_anchor=(1.8,0.5))
plt.show()
```
